Below you will find code to run a very simple Deep Learning example for training a model on the IMDB Reviews dataset. 

For further details have a look at the Tensorflow website from which this is based:
https://www.tensorflow.org/tutorials/keras/text_classification_with_hub

```python
import os
import numpy as np
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_datasets as tfds

# Download the data into training, validation and test. 
train_data, validation_data, test_data = tfds.load(
    name="imdb_reviews", 
    split=('train[:60%]', 'train[60%:]', 'test'),
    as_supervised=True)

def run_training(model):
  # Compile and train the model 
  model.compile(optimizer='adam',
                loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                metrics=['accuracy'])

  history = model.fit(train_data.shuffle(10000).batch(512),
                      epochs=10,
                      validation_data=validation_data.batch(512),
                      verbose=0)

  # Evaluate on test data 
  results = model.evaluate(test_data.batch(512), verbose=2)

  return results 

# Load a pre-trained embedding layer 
embedding = "https://tfhub.dev/google/nnlm-en-dim50/2"
hub_layer = hub.KerasLayer(embedding, input_shape=[], 
                           dtype=tf.string, trainable=True)

# Build the model including the embedding layer
model = tf.keras.Sequential()
model.add(hub_layer)
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(1))

# Print and summarise the model
model.summary()

run_training(model)
```
